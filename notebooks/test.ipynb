```python
import sys
import random

sys.path.append('..')

from src.tasks import DelayDiscountTask
```

```python
task = DelayDiscountTask(100, ['1 week', '2 weeks', '1 month', '2 months', '6 months', '1 year', '5 years', '25 years'], 8)
```

```python
is_running = task.get_is_running()

while is_running:
    choice_1, choice_2 = task.get_trial()
    print(f"{choice_1} or {choice_2}?")

    valid_response = False
    while not valid_response:
        print("Enter your choice:", end=' ')
        choice = random.choice([1, 2])
        print(choice)
        valid_response = task.record_response(choice)

    is_running = task.get_is_running()
```
